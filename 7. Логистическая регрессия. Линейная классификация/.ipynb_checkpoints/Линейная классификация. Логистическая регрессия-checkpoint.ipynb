```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10,10,100)
y = X * (np.random.random_sample(len(X)) + 0.5)
```

```python
X = X.reshape(len(X),1)
```

```python
plt.scatter(X,y);
```

```python
def compute_cost(X, y, theta):
    m = len(y)

    cost = (1/m) * (np.linalg.norm(X @ theta - y) ** 2)
    return cost
```

```python
def stochastic_gradient_descent(X, y, learning_rate, iterations):

    X = np.hstack((np.ones((X.shape[0],1)), X))
    params = np.random.rand(X.shape[1])

    cost_track = np.zeros((iterations, 1))

    for i in range(iterations):
        ind = random.sample(range(X.shape[0]),1)

        params = params - learning_rate * 2 * (X[ind] * ((X[ind] @ params) - y[ind]))
        params = params.ravel()
        cost_track[i] = compute_cost (X, y, params)
    return cost_track, params
```

```python
def predict(X, params):
    X = np.hstack((np.ones((X.shape[0],1)), X))
    return X @ params
```

```python
track, weights = stochastic_gradient_descent(X, y, 0.01, 100)
```

```python
pred = predict(X, weights)
```

```python
plt.scatter(X, y)
plt.plot(X, pred, '-', c = 'r');
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
import numpy as np
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = load_breast_cancer()

X = data.data
y = data.target

X.shape, y.shape
```

```python
len(y[y==0]), len(y[y==1])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
model = LogisticRegression()
model.fit(Xtrain,ytrain)
```

```python
pred_test = model.predict(Xtest)

accuracy_score(ytest, pred_test)
```

```python
pred_test[:5]
```

```python
probs = model.predict_proba(Xtest)

probs[:5]
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold

best_score = -1
best_C = 0

cv = StratifiedKFold(n_splits=5) #   

for C in np.arange(0.1, 10.1, 0.1):
    model = LogisticRegression(C=C)
    score = cross_val_score(model, Xtrain, ytrain, scoring='accuracy', cv=cv).mean()
    if score > best_score:
        best_score = score
        best_C = C
        
print('best score:', best_score)
print('best params:', best_C)
```

```python
model2 = LogisticRegression(C=10.0)
model2.fit(Xtrain, ytrain)

pred_test2 = model2.predict(Xtest)


accuracy_score(ytest,pred_test2)
```

```python
best_score = -1
best_n = 0

cv = StratifiedKFold(n_splits=5) #   

for n in np.arange(1, 22):
    model = KNeighborsClassifier(n_neighbors=n)
    score = cross_val_score(model, Xtrain, ytrain, scoring='accuracy', cv=cv).mean()
    if score > best_score:
        best_score = score
        best_n = n
        
print('best score:', best_score)
print('best params:', best_n)
```

```python
model4 = KNeighborsClassifier(n_neighbors=15)

model4.fit(Xtrain, ytrain)

pred_test4 = model4.predict(Xtest)

accuracy_score(ytest, pred_test4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Murcha1990/Skillbox/master/credit_scoring.csv')
```

```python
df.head()
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix
```

```python
print(':', precision_score(df['target'], df['pred_classes']),
      ':', recall_score(df['target'], df['pred_classes'])
     )
```

```python
import itertools
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment='center',
                 color='white' if cm[i,j] > thresh else 'black')

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
plot_confusion_matrix(confusion_matrix(df['target'],df['pred_classes']),classes=['0', '1'])
```

```python
for threshold in np.arange(0.1, 1, 0.2): #     
    classes = (df['pred_probs'] > threshold).astype(int)

    print('\:', threshold)
    plot_confusion_matrix(confusion_matrix(df['target'], classes), classes=['0','1'])
    print(':', precision_score(df['target'], classes),
      ':', recall_score(df['target'], classes)
     )
    show()
```
