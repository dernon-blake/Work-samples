# Задача

Один из способов повысить эффективность взаимодействия банка с клиентами — отправлять предложение о новой услуге не всем клиентам, а только некоторым, которые выбираются по принципу наибольшей склонности к отклику на это предложение.

Задача заключается в том, чтобы предложить алгоритм, который будет выдавать склонность клиента к положительному или отрицательному отклику на предложение банка. Предполагается, что, получив такие оценки для некоторого множества клиентов, банк обратится с предложением только к тем, от кого ожидается положительный отклик.


Для решения этой задачи загрузите дамп базы в Postgres.

```python
# подключить необходимые библиотеки
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  confusion_matrix
from sklearn.metrics import precision_score, recall_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score

import warnings
warnings.filterwarnings('ignore')
# Ваш код здесь
```

```python
# считать данные
df_clients = pd.read_csv('d_clients.csv')
df_clients = df_clients.set_index("id")
df_agreement = pd.read_csv('d_agreement.csv')


df_agreement_clients = df_agreement.merge(
    df_clients,
    how = "left",
    left_on = "id_client",
    right_index = True
)
df_agreement_clients.head()
# Ваш код здесь
```

```python
df_salary = pd.read_csv('d_salary.csv')
df_agreement_clients = df_agreement_clients.merge(df_salary, how = "inner", on = "id_client")
df_agreement_clients.head()
```

```python
df_close_loan = pd.read_csv('d_close_loan.csv')
df_loan = pd.read_csv('d_loan.csv')
df_loan = df_loan.merge(df_close_loan, how = "inner", on = "id_loan")
df_loan.head()
```

```python
# Добавляем поле с закрытыми сделками каждого клиента
df_clients_closed_fl = df_loan.groupby(['id_client'])['closed_fl'].sum().reset_index(name='loan_num_closed')
df_clients_closed_fl.head()
```

```python
# Добавляем поле с общим кол-вом сделок клиента
df_clients_loan = df_loan.groupby(['id_client'])['id_loan'].count().reset_index(name='loan_num_total')
df_clients_loan.head()
```

```python
df_clients_loan = df_clients_loan.merge(df_clients_closed_fl, how = "inner", on = "id_client")
df = df_agreement_clients.merge(df_clients_loan, how = "inner", on = "id_client")
df
```

```python
# Редактируем некоторые фичи в более удобный вид
print(df.socstatus_work_fl.unique())
df.loc[df['socstatus_work_fl'] == 2, 'socstatus_work_fl'] = 0
df.socstatus_work_fl.unique()
```

```python
print(df.socstatus_pens_fl.unique())
df.loc[df['socstatus_pens_fl'] == 2, 'socstatus_pens_fl'] = 0
df.socstatus_pens_fl.unique()
```

```python
df = df.drop(['reg_address_province', 'postal_address_province', 'family_income', 'fact_address_province'], axis=1)
df.head()
```

```python
# Анализ фичи "семейнео положение"

print(df.marital_status.unique())

# Для снижения вариабельности переменной marital_status, заменим значения на:
#   'Состою в браке', 'Гражданский брак' на 1 (есть партнер)
#   'Разведен(а)', 'Не состоял в браке', 'Вдовец/Вдова' на 0 (партнер отсутствует)

level_map = {'Состою в браке': 1, 'Гражданский брак': 1, 'Разведен(а)': 0, 'Не состоял в браке': 0, 'Вдовец/Вдова': 0}
df['marital_status'] = df['marital_status'].map(level_map)
df.marital_status.unique()
```

```python
# проведите анализ данных, посмотрите на индивидуальные графики
# зависимости целевой функции и отдельной переменной
# создайте дополнительные переменные
# для работы с категориальными данными можно воспользоваться pd.get_dummies
# сделайте предварительные выводы

# Ваш код здесь

df = pd.read_csv('bank_interaction_clients.csv')
df.head()
```

```python
sns.pairplot(df, hue='target');
```

```python
df = pd.get_dummies(data=df, dtype=int)
df.head()
```

```python
df.corr()
```

```python
# раздедлите на обучающую и тестовую выборки
train, test = train_test_split(df, train_size=0.7, random_state=42)

train_X = train.drop(['target'], axis=1)
train_y = train['target']

test_X = test.drop(['target'], axis=1)
test_y = test['target']

# Ваш код здесь
```

```python
# обучите модель методом ближайших соседей
score = []
for i in range(1, 200, 2):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(train_X, train_y)

    score_train = knn.score(train_X, train_y)
    score_test = knn.score(test_X, test_y)

    score.append([i, score_train, score_test])
    
score        
# Ваш код здесь
```

```python
# проверьте качество модели с помощью F-мера
score_df = pd.DataFrame(score, columns = ['i', 'score_train', 'score_test'])
score_df[score_df['score_test'] == max(score_df['score_test'])]
# Ваш код здесь
```

```python
loo = LeaveOneOut()
```

```python
# проверьте качество модели с помощью матрицы сопряжённостей
score = []
for i in range(1,50,2):
    model_knn = KNeighborsClassifier(n_neighbors=i)
    scores = cross_val_score(model_knn, train_X, train_y,
                         scoring='accuracy',
                         cv=loo, n_jobs = -1)
    score.append([i, scores.mean()])

score_df = pd.DataFrame(score, columns =['i','score'])
score_df[score_df['score'] == score_df['score'].max()]
# Ваш код здесь
```

```python
model_knn = KNeighborsClassifier(n_neighbors=23)
model_knn.fit(train_X, train_y)
confusion_matrix(test_y, model_knn.predict(test_X))
```

```python
report = classification_report(test_y, model_knn.predict(test_X), target_names=['0', '1'])
print(report)
```
