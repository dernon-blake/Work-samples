```python
def borders(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3 - q1
    boundaries = (q1 - iqr * 1.5, q3 + iqr * 1.5)
    return boundaries

boundaries = borders(df_out.price)
is_outlier_price = (df_out.price < boundaries[0]) | (df_out.price > boundaries[1])
```

```python
df_out = df_out[~is_outlier_price]
```

### Feature engineering

```python
# Создание дополнительных признаков
df_out['short_model'] = df_out.apply(lambda x: x.model.split(' ')[0], axis = 1)
df_out['month']=df_out.date.dt.month
df_out['dayofweek']=df_out.date.dt.dayofweek
df_out['diff_years']=df_out.apply(lambda x: x.date.year-x.year, axis = 1)
df_out['age_category'] = df_out.apply(lambda x: 'new' if x.year > 2016 else('old' if x.year < 2009 else 'average'), axis=1)
```

```python
# Преобразуем категориальные переменные
column_trans = ColumnTransformer(
    [('categories', OneHotEncoder(dtype='int',sparse_output=False), [
        'fuel','transmission','region','state','short_model','manufacturer','title_status','age_category'])],
     remainder='drop', verbose_feature_names_out=False)
column_trans.fit(df_out)
col_names = column_trans.get_feature_names_out().tolist()
df_out[col_names] = column_trans.transform(df_out)
```

```python
# Стандартизируем числовые данные
std = StandardScaler()
new_names = ['odometer_std','lat_std','long_std','year_std','month_std','dayofweek_std','diff_years_std']
std.fit(df_out[['odometer','lat','long','year','month','dayofweek','diff_years']])
df_out[new_names]=std.transform(df_out[['odometer','lat','long','year','month','dayofweek','diff_years']])
```

```python
# Убираем ненужные колонки и получаем финальный датафрейм
columns_for_drop = ['year', 'url', 'region', 'region_url', 'manufacturer',
                    'model', 'fuel', 'odometer', 'title_status', 'transmission',
                    'image_url', 'description', 'state', 'lat', 'long', 'posting_date',
                    'age_category','short_model','date',
                    'month', 'dayofweek', 'diff_years'
                    ]
df_prepared = df_out.drop(columns = columns_for_drop)
df_prepared.head()
```

```python
# Сохраняем финальный датафрейм для разработки модели
df_prepared.to_csv('auto_dataset_prepared.csv',index=False)
```

```python
# По результатам кросс-валидации лучшей моделью стала логистическая регрессия. Но т.к. разница у нее и моделью
# случайного леса минимальна, а также, при инициализации модели вылазит ошибка, использую модель случайного леса.
rf_tuned.fit(x_train,y_train)
rf_pred = rf_tuned.predict(x_test)
print(f'Точность на тренировочной выборке: {accuracy_score(rf_tuned.predict(x_train),y_train)}')
print(f'Точность на тестовой выборке: {accuracy_score(rf_pred,y_test)}')
confusion_matrix(rf_pred,y_test)
```
