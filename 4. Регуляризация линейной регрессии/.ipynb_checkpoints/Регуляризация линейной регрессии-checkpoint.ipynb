```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(0)
x = np.random.rand(100,1) * 10
y = 1000 + 7 * np.sin(x) + 2 * np.random.randn(100,1)

d = {'x': x.reshape(1,100)[0],
     'y': y.reshape(1,100)[0]}
df = pd.DataFrame(data = d)
df['y'] = df ['y'] * 10
df.head()
```

```python
plt.scatter(df['x'], df['y'], s=10)
df = df.sort_values('x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
df['x2'] = df['x'].apply(lambda x: np.sin(x))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
train, test = train_test_split(df, train_size=0.7, random_state=123)
```

```python
model = LinearRegression()
model.fit (train[['x','x2']], train[['y']])
```

```python
train = train.sort_values('x')
```

```python
y_predict_train = model.predict(train[['x','x2']])
y_predict_test = model.predict(test[['x','x2']])
```

```python
plt.scatter(train['x'], train['y'], s=10, c='b')
plt.scatter(test['x'], test['y'], s=10, c='r')
plt.plot(train['x'], y_predict_train, c='g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

train_mse = mean_squared_error(train[['y']], y_predict_train)
test_mse = mean_squared_error(test[['y']], y_predict_test)
print('MSE train:', train_mse)
print('MSE test:', test_mse)
```

```python
train_mae = mean_absolute_error(train[['y']], y_predict_train)
test_mae = mean_absolute_error(test[['y']], y_predict_test)
print('MAE train:', train_mae)
print('MAE test:', test_mae)
```

```python
train['y_predict_train'] = y_predict_train
```

```python
residual = train['y'] - train['y_predict_train']
```

```python
plt.scatter(train['y_predict_train'], train['y'], c ='g')
plt.xlabel('предсказанное значение y_predict')
plt.ylabel('реальное значение y')
plt.show()
```

```python
plt.scatter(train['y_predict_train'], residual, c ='g')
plt.xlabel('предсказанное значение y_predict')
plt.ylabel('значение остатков')
plt.show()
```

```python
from sklearn import preprocessing
normalized_residual = preprocessing.normalize([residual])
```

```python
plt.scatter(train['y_predict_train'], normalized_residual, c ='g')
plt.xlabel('предсказанное значение y_predict')
plt.ylabel('значение остатков')
plt.show()
```

```python
train['y_predict_train2'] = y_predict_train2
residual2 = train['y'] - train['y_predict_train2']

plt.scatter(list(range(len(train['y_predict_train2']))), residual2, c='g')
plt.xlabel('номер наблюдения')
plt.ylabel('значения остатков')
plt.show ()
```

```python
from sklearn.metrics import r2_score
```

```python
residual = train['y'] - train['y_predict_train']
total = train['y'] - train['y'].mean()

train_r2 = 1 - sum(residual.apply(lambda x: x**2))/sum(total.apply(lambda x: x**2))
print('R2 train:', train_r2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df.head()
```

```python
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(x=df['YearsExperience'], y=df['Salary'])
plt.show()
```

```python
plt.xlabel('Floor')
plt.ylabel('Salary')
plt.scatter(x=df['Floor'], y=df['Salary'])
plt.show()
```

```python
plt.xlabel('Type')
plt.ylabel('Salary')
plt.scatter(x=df['Type'], y=df['Salary'])
plt.show()
```

```python
df['x'] = df.Floor ** 2
```

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=3)
```

```python
train_df = train_df.sort_values(by='YearsExperience', ascending=False)
train_X = train_df[['YearsExperience', 'Floor', 'Type', 'x']]
train_y = train_df [['Salary']]

test_df = test_df.sort_values(by='YearsExperience', ascending=False)
test_X = test_df[['YearsExperience', 'Floor', 'Type', 'x']]
test_y = test_df [['Salary']]
```

```python
linear_regression = LinearRegression()
linear_regression.fit(train_X,train_y)
```

```python
print(linear_regression.score(train_X,train_y))
print(linear_regression.score(test_X,test_y))
```

```python
print('w_0:', linear_regression.intercept_) #intercept_ - метод на вес w0 параметра
coef = list(linear_regression.coef_[0]) # coef_[i] - метод на весь wi параметра 
for i in range(len(coef)):
    print(f'w_{i+1}:', coef[i], train_X.columns[i])
```

```python
y_predict_train = linear_regression.predict(train_X)

plt.scatter(train_X.YearsExperience, train_y, s=10, c ='b')
plt.scatter(test_X.YearsExperience, test_y, s=10, c='r')
plt.plot(train_X.YearsExperience, y_predict_train, c='g')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_regression = Lasso(alpha=0.1)
lasso_regression.fit(train_X,train_y)
```

```python
print(lasso_regression.score(train_X,train_y))
print(lasso_regression.score(test_X,test_y))
```

```python
print('w_0:', lasso_regression.intercept_)
print('w_i:', lasso_regression.coef_)
```

```python
coef = []
for i in range(-4,5):
    alpha = 10**i
    lasso_regression = Lasso(alpha=alpha)
    lasso_regression.fit(train_X, train_y)

    coef.append(list(lasso_regression.coef_))

    print('-'*5, 'lambda =', alpha, '-'*5)
    print(lasso_regression.score(train_X, train_y))
    print(lasso_regression.score(test_X,test_y))
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
from sklearn.linear_model import Ridge
```

```python
df['x'] = df['Floor'] ** 2
train_df, test_df = train_test_split(df, test_size=0.3, random_state=3)
train_df = train_df.sort_values(by='YearsExperience', ascending=False)
train_X = train_df[['YearsExperience', 'Floor', 'Type', 'x']]
train_y = train_df [['Salary']]

test_df = test_df.sort_values(by='YearsExperience', ascending=False)
test_X = test_df[['YearsExperience', 'Floor', 'Type', 'x']]
test_y = test_df [['Salary']]
```

```python
y_predict_train = linear_regression.predict(train_X)

plt.scatter(train_X.YearsExperience, train_y, s=10, c ='b')
plt.scatter(test_X.YearsExperience, test_y, s=10, c='r')
plt.plot(train_X.YearsExperience, y_predict_train, c='g')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
ridge_reg = Ridge(alpha=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
print(ridge_reg.score(train_X,train_y))
print(ridge_reg.score(test_X,test_y))
```

```python
print('w_0:', ridge_reg.intercept_)
print('w_i:', ridge_reg.coef_)
```

```python
coef = []
for i in range(-7, -1):
    alpha = 10 ** i
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(train_X, train_y)

    coef.extend(list(ridge_reg.coef_))

    print('-'*5, 'lambda =', alpha, '-'*5)
    print(ridge_reg.score(train_X,train_y))
    print(ridge_reg.score(test_X,test_y))

```

```python
from sklearn import preprocessing # нормализируем данные
```

```python
d = preprocessing.normalize(df, axis=0)
scaled_df = pd.DataFrame(d, columns=df.columns)
scaled_df.head()
```

```python
train_df, test_df = train_test_split(scaled_df, test_size=0.3, random_state=3)
```

```python
train_df = train_df.sort_values(by='YearsExperience', ascending=False)
train_X = train_df[['YearsExperience', 'Floor', 'Type', 'x']]
train_y = train_df [['Salary']]

test_df = test_df.sort_values(by='YearsExperience', ascending=False)
test_X = test_df[['YearsExperience', 'Floor', 'Type', 'x']]
test_y = test_df [['Salary']]
```

```python
coef = []
for i in range(-7, -1):
    alpha = 10 ** i
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(train_X, train_y)

    coef.extend(list(ridge_reg.coef_))

    print('-'*5, 'lambda =', alpha, '-'*5)
    print(ridge_reg.score(train_X,train_y))
    print(ridge_reg.score(test_X,test_y))
```
