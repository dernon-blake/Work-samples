```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

plt.rcParams['figure.figsize'] = (11, 6.5)
```

```python
np.random.seed(13)
n = 500
X = np.zeros(shape=(n, 2))
X[:, 0] = np.linspace(-5,5, 500)
X[:, 1] = X[:,0] + 0.5 * np.random.normal(size=n)
y = (X[:, 1] > X[:, 0]).astype(int)
plt.scatter(X[:,0], X[:, 1], s=100, c=y, cmap='winter')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=13)

lr = LogisticRegression()
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)

accuracy_score(y_pred_lr, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

accuracy_score(y_pred_dt, y_test)
```

Видно, что на линейных данных линейные модели работают лучше, чем теже деревья

```python
np.random.seed(13)
X = np.random.randn(500, 2)
y = np.logical_xor(X[:, 0] >0, X[:, 1] >0).astype(int)
plt.scatter(X[:, 0], X[:, 1], s=100, c=y, cmap='autumn')
plt.show();
#КУСОЧНО-ЛИНЕЙНЫЕ ДАННЫЕ
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)

accuracy_score(y_pred_dt, y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris

data = load_iris(as_frame=True)

X = data.data[['sepal length (cm)', 'sepal width (cm)']]
y = data.target
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=45)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)

y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
tree.get_depth()
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred), accuracy_score(y_test,y_test_pred))
```

```python
path = tree.cost_complexity_pruning_path(X_train, y_train)
alphas = path['ccp_alphas']

alphas
```

```python
accuracy_train, accuracy_test = [], []

for i in alphas:
    tree = DecisionTreeClassifier(ccp_alpha = i)

    tree.fit(X_train, y_train)
    y_train_pred = tree.predict(X_train)
    y_test_pred = tree.predict(X_test)

    accuracy_train.append(accuracy_score(y_train, y_train_pred))
    accuracy_test.append(accuracy_score(y_test, y_test_pred))

sns.set()
plt.figure(figsize=(14,7))
sns.lineplot(y = accuracy_train, x = alphas, label = "Train Accuracy")
sns.lineplot(y = accuracy_test, x = alphas, label = "Test Accuracy")
plt. xticks(ticks = np.arange(0.00,0.25,0.01))
plt.show();
```

```python
tree = DecisionTreeClassifier(ccp_alpha = 0.02, random_state=40)
tree.fit(X_train, y_train)

y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)

print(accuracy_score(y_train, y_train_pred), accuracy_score(y_test, y_test_pred))
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

data = fetch_california_housing(as_frame=True)
```

```python
X = data.data
y = data.target

X.head()
```

```python
X.isna().sum() # пропуски
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X,y, test_size=0.25, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor()

model_dt.fit(Xtrain, ytrain)
```

```python
pred_dt = model_dt.predict(Xtest)

from sklearn.metrics import mean_squared_error

mean_squared_error(ytest, pred_dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor()

model_rf.fit(Xtrain,ytrain)
```

```python
pred_rf = model_rf.predict(Xtest)

mean_squared_error(ytest, pred_rf) # ошибка в 2 раза меньше, чем у дерева
```

```python
N = []
MSE = []

for n in np.arange(10, 120, 20):
    rf = RandomForestRegressor(n_estimators = n)

    rf.fit(Xtrain, ytrain)

    pred_rf = rf.predict(Xtest)

    N.append(n)
    MSE.append(mean_squared_error(ytest, pred_rf))
```

```python
from matplotlib import pyplot as plt

plt.plot(N, MSE, '-')
plt.xlabel('numbers of trees in forest')
plt.ylabel('MSE on test');
```

```python
D = []
MSE = []

for d in np.arange(2, 50, 5):
    rf = RandomForestRegressor(n_estimators = 100, max_depth=d)

    rf.fit(Xtrain, ytrain)

    pred_rf = rf.predict(Xtest)

    D.append(d)
    MSE.append(mean_squared_error(ytest, pred_rf))
```

```python
plt.plot(D, MSE, '-')
plt.xlabel('max_depth of trees')
plt.ylabel('MSE on test');
```

```python
model_rf = RandomForestRegressor(n_estimators=100, max_depth=20)
model_rf.fit(Xtrain, ytrain)

pred_rf = model_rf.predict(Xtest)

mean_squared_error(ytest, pred_rf)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(RandomForestRegressor(n_estimators=100, max_depth=20), X, y, cv=3, scoring='neg_mean_squared_error')
```

```python
scores
```

```python
np.mean (-scores)
```

```python
np.std(-scores)
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np

data = fetch_california_housing(as_frame=True)
```

```python
X = data.data
y = data.target

X.head()
```

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor()

model_rf.fit(Xtrain, ytrain)
```

```python
from sklearn.metrics import mean_squared_error

pred_rf = model_rf.predict(Xtest)

mean_squared_error(ytest, pred_rf)
```

```python
model_rf = RandomForestRegressor(n_estimators = 150)

model_rf.fit(Xtrain, ytrain)

pred_rf = model_rf.predict(Xtest)

mean_squared_error(ytest, pred_rf)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'max_depth': np.arange(2, 75, 4),
          'max_features': ['sqrt', 'log2', None]}

gs = GridSearchCV(RandomForestRegressor(n_estimators=150), params, cv=3, scoring='neg_mean_squared_error')

gs.fit(Xtrain, ytrain)
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

```python
model = gs.best_estimator_

model.fit(Xtrain, ytrain)
```

```python
pred = model.predict(Xtest)

mean_squared_error(ytest, pred)
```
