```python
import pandas as pd
import requests
```

```python
url = 'http://itunes.apple.com/search'
params = {'media':'music', 'term':'manizha', 'limit': 30}
response = requests.get(url, params=params)
response
```

```python
response.status_code
```

```python
response.json ()
```

```python
results = response.json()['results']
results
```

```python
df = pd.DataFrame(results)
```

```python
df
```

```python
df.primaryGenreName.value_counts()
```

```python
help(df.to_excel)
```

```python
df.to_excel('data/iTunes_api.xlsx', sheet_name='Manizha', index_label='id')
```

```python
df.to_csv('data/iTunes_api.csv', sep = '\t', index = False)
```

# POST

```python
url = 'http://itunes.apple.com/search'
params = {'media':'music', 'term':'manizha', 'limit': 30}
response = requests.post(url, data=params, timeout = 10)
response
#ТАК КАК ПРИ POST МЫ ПАРАМЕТРЫ ЗАДЕМ НЕ ЧЕРЕЗ АДРЕСНОЮ СТРОКУ, А ЧЕРЕЗ ТЕЛО ЗАПРОСА, ТО PARAMS МЕНЯЕТСЯ НА DATA.
#timeout - время ожидания ответа с сервера в секундах. Если ответ идет дольше заявленного времени, то связь обрывается, даже без ответа.
```

```python
response.json ()
```
